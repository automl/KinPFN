```python
import pandas as pd

data_tRNA = pd.read_csv("./tRNA_kinfold_timing.csv")
data_tRNA_short = pd.read_csv("./tRNA_kinfold_timing_25_50_1000.csv")
data_rRNA = pd.read_csv("./rRNA_kinfold_timing.csv")

tRNA_runtime_dict = {
    "num_simulations": [],
    "avg_total_computation_time": [],
}

tRNA_runtime_dict_short = {
    "num_simulations": [],
    "avg_total_computation_time": [],
}

rRNA_runtime_dict = {
    "num_simulations": [],
    "avg_total_computation_time": [],
}

for i in range(0, 5):
    tRNA_runtime_dict["num_simulations"].append(data_tRNA["num_simulations"][i])
    tRNA_runtime_dict["avg_total_computation_time"].append(
        data_tRNA["avg_total_computation_time"][i]
    )
    

    rRNA_runtime_dict["num_simulations"].append(data_rRNA["num_simulations"][i])
    rRNA_runtime_dict["avg_total_computation_time"].append(
        data_rRNA["avg_total_computation_time"][i]
    )

for i in range(0, 2):  
    tRNA_runtime_dict_short["num_simulations"].append(data_tRNA_short["num_simulations"][i])
    tRNA_runtime_dict_short["avg_total_computation_time"].append(
        data_tRNA_short["avg_total_computation_time"][i]
    )  


df_tRNA = pd.DataFrame(tRNA_runtime_dict)
df_tRNA_short = pd.DataFrame(tRNA_runtime_dict_short)
df_rRNA = pd.DataFrame(rRNA_runtime_dict)

df_tRNA["RNA"] = "tRNA^phe"
df_tRNA_short["RNA"] = "tRNA^phe"
df_rRNA["RNA"] = "5S-rRNA"

df_combined = pd.concat([df_tRNA, df_tRNA_short,df_rRNA])
```

```python
import matplotlib.pyplot as plt
import numpy as np

num_simulations = [10, 25, 50, 75, 1000]
avg_total_cpu_time_tRNA_minutes = df_tRNA["avg_total_computation_time"] / 60
avg_total_cpu_time_rRNA_minutes = df_rRNA["avg_total_computation_time"] / 60


bar_width = 0.35
gap = 0.02

index = np.arange(len(num_simulations))

fig, ax = plt.subplots(figsize=(5.70, 3.5))
fig.set_dpi(600)

ax.grid(
    True, which="both", axis="y", color="gray", linestyle="--", linewidth=0.5, zorder=1
)

bar1 = ax.bar(
    index - (bar_width + gap) / 2,
    avg_total_cpu_time_tRNA_minutes,
    bar_width,
    label=r"tRNA$^{\text{phe}}$",
    color="darkblue",
    zorder=2,
)
bar2 = ax.bar(
    index + (bar_width + gap) / 2,
    avg_total_cpu_time_rRNA_minutes,
    bar_width,
    label="5S-rRNA",
    color="darkred",
    zorder=2,
)


ax.set_xlabel("Number of Folding Simulations")
ax.set_ylabel(r"Kinfold CPU Time (Minutes)")
ax.set_xticks(index)
ax.set_xticklabels(num_simulations)
ax.legend()

plt.tight_layout()
plt.show()
```
